```python
# 获取talib的全部动量指标
class Index:
    def __init__(self):
        self.index_list = ["ADX, ADXR","APO", "AROON_aroondown", "AROONOSC", "BOP","CCI",
                          "CMO", "DX", "MACD_macd", "MACDEXT_macd", "MACDFIX_macd","MFI",
                          "MINUS_DI", "MINUS_DM", "MOM", "PLUS_DI", "PLUS_DM", "PPO", "ROC",
                          "ROCP", "ROCR", "RSI","STOCH", "STOCHF", "TRIX", "ULTOSC", "WILLR", "VIXFIX"]
    def Momentum_Indicator(self, df, indicator):
        """
        df: dataframe, 股票数据
        indicator: str, 指标
        """
        open_p = df['Open'].values
        high_p = df['High'].values
        close_p = df['Close'].values
        low_p = df['Low'].values
        volume = df["Volume"].values
        volume_p = df["Volume"].values

        float_data = [float(x) for x in df["Volume"].values]     
        volume_p = np.array(float_data)
        if indicator == "ADX":
            df["ADX"] = talib.ADX(high_p, low_p, close_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "ADXR":
            df["ADXR"] = talib.ADXR(high_p, low_p, close_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "APO":
            df["APO"] = talib.APO(close_p, fastperiod=12, slowperiod=26, matype=0)
            df = self.normalise(df, indicator)
            return df
        if indicator == "AROON_aroondown":
            df["AROON_aroondown"],df["AROON_aroonup"] = talib.AROON(high_p, low_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "AROONOSC":
            df["AROONOSC"] = talib.AROONOSC(high_p, low_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "BOP":
            df["BOP"] = talib.BOP(open_p, high_p, low_p, close_p)
            df = self.normalise(df, indicator)
            return df 
        if indicator == "CCI":    
            df["CCI"] = talib.CCI(high_p, low_p, close_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "CMO":
            df["CMO"] = talib.CMO(close_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df 
        if indicator == "DX":
            df["DX"] = talib.DX(high_p, low_p, close_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "MACD_macd":
            df["MACD_macd"],df["MACD_macdsignal"],df["MACD_macdhist"] = talib.MACD(close_p, fastperiod=12, slowperiod=26, signalperiod=9)
            df = self.normalise(df, indicator)
            return df
        if indicator == "MACDEXT_macd":
            df["MACDEXT_macd"],df["MACDEXT_macdsignal"],df["MACDEXT_macdhist"] = talib.MACDEXT(close_p, fastperiod=12, fastmatype=0, slowperiod=26, slowmatype=0, signalperiod=9, signalmatype=0)
            df = self.normalise(df, indicator)
            return df
        if indicator == "MACDFIX_macd":
            df["MACDFIX_macd"],df["MACDFIX_macdsignal"],df["MACDFIX_macdhist"] = talib.MACDFIX(close_p, signalperiod=9)
            df = self.normalise(df, indicator)
            return df
        if indicator == "MFI":
            df["MFI"] = talib.MFI(high_p, low_p, close_p, volume_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "MINUS_DI":
            df["MINUS_DI"] = talib.MINUS_DI(high_p, low_p, close_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "MINUS_DM":
            df["MINUS_DM"] = talib.MINUS_DM(high_p, low_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "MOM":
            df["MOM"] = talib.MOM(close_p, timeperiod=10)
            df = self.normalise(df, indicator)
            return df
        if indicator == "PLUS_DI":
            df["PLUS_DI"] = talib.PLUS_DI(high_p, low_p, close_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "PLUS_DM":
            df["PLUS_DM"] = talib.PLUS_DM(high_p, low_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "PPO":
            df["PPO"] = talib.PPO(close_p, fastperiod=12, slowperiod=26, matype=0)
            df = self.normalise(df, indicator)
            return df
        if indicator == "ROC":
            df["ROC"] = talib.ROC(close_p, timeperiod=10)
            df = self.normalise(df, indicator)
            return df
        if indicator == "ROCP":
            df["ROCP"] = talib.ROCP(close_p, timeperiod=10)
            df = self.normalise(df, indicator)
            return df
        if indicator == "ROCR":
            df["ROCR"] = talib.ROCR(close_p, timeperiod=10)
            df = self.normalise(df, indicator)
            return df
        if indicator == "RSI":
            df["RSI"] = talib.RSI(close_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "STOCH":
            df["STOCH_slowk"],df["STOCH_slowd"] = talib.STOCH(high_p, low_p, close_p, fastk_period=5, slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)
            df = self.normalise(df, indicator)
            return df
        if indicator == "STOCHF":
            df["STOCHF_fastk"],df["STOCHF_fastd"] = talib.STOCHF(high_p, low_p, close_p, fastk_period=5, fastd_period=3, fastd_matype=0)
            df = self.normalise(df, indicator)
            return df
        if indicator == "TRIX":
            df["TRIX"] = talib.TRIX(close_p, timeperiod=30)
            df = self.normalise(df, indicator)
            return df
        if indicator == "ULTOSC":
            df["ULTOSC"] = talib.ULTOSC(high_p, low_p, close_p, timeperiod1=7, timeperiod2=14, timeperiod3=28)
            df = self.normalise(df, indicator)
            return df
        if indicator == "WILLR":
            df["WILLR"] = talib.WILLR(high_p, low_p, close_p, timeperiod=14)
            df = self.normalise(df, indicator)
            return df
        if indicator == "VIXFIX":
            df = self.VIXFIX(df)
            df = self.normalise(df, indicator)
            return df
        return df

    def normalise(self, data, indicator):
        """
        标准化每一个指标
        """
        for column in data.columns:
            if indicator in column:
                df = data[~np.isnan(data[column])]
                df = df[~np.isinf(df[column])][column]
                avg, std = norm.fit(df)
                norm_list = []
                for i in range(data.shape[0]):
                    if np.isnan(data[column][i]):
                        norm_list.append(0)
                    else:
                        zscore = (data[column][i]-avg) /std
                        norm_list.append(norm.cdf(zscore) * 100)
                # 标准化
                norm_indicator = column + "_NORM"
                data[norm_indicator] = norm_list
        return data
    def VIXFIX(self, data):
        """
        Function is to calculate vixfix index based on the input data.
        """
        if "Close" not in data.columns:
            return data
        def HHV(data, N):
            data["HHV"] = data.Close.rolling(N).max().shift()
            data = data.fillna(0)
            return data
        data = HHV(data, N=21)
        # VIXFIX 未标准化
        data["VIXFIX"] = 100 * (data["HHV"] - data["Low"]) / data["HHV"]
        return data 
```

```python
import pandas as pd
import os
import numpy as np
import talib 
from talib import MA_Type
import pandas as pd
import datetime
import os
from urllib.request import urlopen
import json
import itertools
import scipy
from numpy import inf
from scipy.stats import norm
import pyecharts.options as opts
from pyecharts.charts import Candlestick,Line
def test(indicator, time_period, buy_thresholds, sell_thresholds, test_file_list):
    """
    parameters:
    indicator: str
    time_period: int, 回测天数
    buy_thredhold: 列表, 记录买入阈值
    sell_threshold: 列表, 记录卖出阈值
    """
    data_path = "/Users/liulu/Desktop/stock/test/data/SP_500/"
    # 输入测试数据
    buy_y_rate = {}
    sell_y_rate = {}
    comb_threshold = list(itertools.product(buy_thresholds,sell_thresholds))
    for buy_threshold, sell_threshold in comb_threshold:
        # 计算不同阈值的成功率
        buy = {}
        sell = {}
        for file in test_file_list:
            stock_name = file[0: file.index(".")]
            path = data_path + file
            # 给出信号
            df = pd.read_csv(path)
            # 调用指标
            index = Index()
            df = index.Momentum_Indicator(df, indicator)
            transform_df = give_signal(df, buy_threshold, sell_threshold)
            buy_signal_total = 0
            buy_signal_y = 0
            buy_signal_n = 0
            sell_signal_total = 0
            sell_signal_y = 0
            sell_signal_n = 0
            for i in range(transform_df.shape[0] - time_period):
                current = transform_df.iloc[i].to_dict()
                future = transform_df.iloc[i+time_period].to_dict()
                if current["signal"] == "buy":
                    buy_signal_total += 1
                    if (future["Close"] - current["Close"])/ current["Close"] > 0:
                        # 买入信号有效
                        buy_signal_y += 1
                    else:
                        buy_signal_n += 1
                
                if current["signal"] == "sell":
                    sell_signal_total += 1
                    if (future["Close"] - current["Close"])/ current["Close"] < 0:
                        # 卖出信号有效
                        sell_signal_y += 1
                    else:
                        sell_signal_n += 1
    
            # 计算成功率
            stock_buy_y_rate = buy_signal_y/buy_signal_total
            stock_sell_y_rate = sell_signal_y/sell_signal_total
            buy[stock_name] = stock_buy_y_rate
            sell[stock_name] = stock_sell_y_rate
        buy_y_rate[(buy_threshold, sell_threshold)] = np.mean(list(buy.values()))
        sell_y_rate[(buy_threshold, sell_threshold)]= np.mean(list(sell.values()))
    return  buy_y_rate, sell_y_rate 
```

```python
def save_to_json(dic, filename):
    newdict = {}
    for day,value in dic.items():
        value_str = {}
        for threshold_comb, rate in value.items():
            value_str[str(threshold_comb)] = rate
        newdict[str(day)] = value_str    
    with open(filename+".json", "w+") as f:
        json.dump(newdict, f)
```

```python
def give_signal(df, buy_threshold, sell_threshold):
    index_columns = [a for a in df.columns if "NORM" in a]
    if index_columns == 1:
        # 只有一个标准化指标
        index = index_columns[0]
        signal = []
        for i, row in df.iterrows():
            if row_todict()[index] <= buy_threshold and row_dict()[index] > 0:
                signal.append("buy")
            elif row_dict()[index] >= sell_threshold:
                signal.append("sell")
            else:
                signal.append(None)
        df["signal"] = signal
    else:
        # 有多个指标,取平均
        signal = []
        for i, row in df.iterrows():
            row = row.to_dict()
            indexes = [row[x] for x in index_columns]
            mean = np.mean(indexes)
            if mean >= buy_threshold:
                signal.append("buy")
            elif mean <= sell_threshold:
                signal.append("sell")
            else:
                signal.append(None)
        df["signal"] = signal
    return df
```

```python
import random 
import json
def main(indicator,data_path):
    """
    indicator: str 测试指标名字
    threshold_parameters: 列表, [buy_threshold, sell_threshold]
    """
    # 导入测试文件路径
    filenames = os.listdir(data_path)
    # 随机选择测试文件
    random.shuffle(filenames)
    testfiles = filenames[:10]
    # 测试
    days =[x for x in range(1, 6)]
    buy_thresholds = [x for x  in range(0, 100)]
    sell_thresholds = [x for x in range(0,100)]
    buy_y_day = {}
    sell_y_day = {}
    for day in days:
        buy_rate, sell_rate = test(indicator, day, buy_thresholds, sell_thresholds, testfiles)
        buy_y_day[day] = buy_rate
        sell_y_day[day] = sell_rate
    save_to_json(buy_y_day, indicator+"-buy")
    save_to_json(sell_y_day, indicator+"-sell")
    buy_points = transform_to_point(buy_y_day, "buy")
    sell_points = transform_to_point(sell_y_day, "sell")
    plot(buy_points, "buy", indicator)
    plot(sell_points, "sell", indicator)
    return 
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pyplot import figure
def plot(points, direction, indicator):
    """
    points: 数据点
    direction: buy or sell
    """
#     max_index = np.argmax(value)
#     day_max = days[max_index]
#     buy_th_max = buy_th[max_index]
#     sell_th_max = sell_th[max_index]
#     value_max = value[max_index]
    days = [x[0] for x in points]
    th = [x[1] for x in points]
    value = [x[2] for x in points]
    max_index = np.argmax(value)
    min_index = np.argmin(value)
    point_max = points[max_index]
    point_min = points[min_index]
    
    fig = plt.figure()
    plt.rcParams["figure.figsize"] = (30,30)
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter3D(days,th,value)
    ax.set_xlabel('day')  # 坐标轴
    ylabel = "{}_th".format(direction)
    ax.set_ylabel(ylabel)
    ax.set_zlabel('successful rate')
    ax.text(point_max[0], point_max[1], round(point_max[2], 3), "max:\n (day={}, {}={}, rate={}".format(point_max[0], ylabel, point_max[1], round(point_max[2], 3)), size=14, ha='center', va='center')
    ax.text(point_min[0], point_min[1], round(point_min[2], 3), "min:\n (day={}, {}={}, rate={}".format(point_min[0], ylabel, point_min[1], round(point_min[2], 3)), size=14, ha='center', va='baseline')
    fig.tight_layout()
    plt.show()
    fig.savefig('{}-{}.png'.format(indicator, direction), dpi=100, bbox_inches='tight')
```

```python
def transform_to_point(dic, direction):
    """
    dic: 字典
    direction: buy or sell
    """
    points = []
    days = dic.keys()
    all_thresholds = list(dic[1].keys())
    for day in days:
        for buy_th, sell_th in all_thresholds:
            value = dic[day][(buy_th, sell_th)]
            if direction == "buy":
                points.append((day, buy_th, value))
            elif direction == "sell":
                points.append((day, sell_th, value))
    return list(set(points))
```
